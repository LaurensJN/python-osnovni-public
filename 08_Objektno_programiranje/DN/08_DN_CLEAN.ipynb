Napišite funkcijo <b>razadalja_v_pomnilniku(prvi_atribut, drugi_atribut)</b>, ki vrne koliko narazen so si objekti v pomnilniku.

Razdalje se bodo razlikovale po vsakem novem zagonu celice.
</div>

Primeri:
```python
Input:
    a = 1
    b = [2,3,4]
    c = {"prvi": 8, "drugi": 9, "tretji":10}
    print(f"a <--- {razadalja_v_pomnilniku(a,b)} ---> b")
    print(f"a <--- {razadalja_v_pomnilniku(a,c)} ---> c")
    print(f"b <--- {razadalja_v_pomnilniku(b,c)} ---> c")

Output:
    # razdalje se bodo razlikovale po vsakem zagonu
    a <--- 72999760 ---> b
    a <--- 90437904 ---> c
    b <--- 17438144 ---> c

```

````python
# Rešitev
def razadalja_v_pomnilniku(first_element, second_element):
    return abs(int(id(first_element)) - int(id(second_element)))


a = 1
b = [2,3,4]
c = {"prvi": 8, "drugi": 9, "tretji":10}
print(f"a <--- {razadalja_v_pomnilniku(a,b)} ---> b")
print(f"a <--- {razadalja_v_pomnilniku(a,c)} ---> c")
print(f"b <--- {razadalja_v_pomnilniku(b,c)} ---> c")
````

````python
# Rešitev
class Zaposlena_oseba:
    def __init__(self, first_name, name, salary):
        self.ime = first_name
        self.priimek = name
        self.placi = salary
        
    def razlika_v_placi(self, person):
        return abs(self.placi - person.placi)
    
zaposleni = [Zaposlena_oseba("Luka", "Novak", 1500.0),
             Zaposlena_oseba("Živa", "Groza", 2100.0),
             Zaposlena_oseba("Anže", "Jaklin", 1436.50),
             Zaposlena_oseba("Luka", "Bogataj", 1790.0),
             Zaposlena_oseba("Sara", "Ljubljen", 3600.0)]

print("Razlika med plačami zaposlenih:")
for i in range(len(zaposleni)):
    for j in range(i+1, len(zaposleni)):
        ime1 = zaposleni[i].ime
        priimek1 = zaposleni[i].priimek
        razlika = zaposleni[i].razlika_v_placi(zaposleni[j])
        ime2 = zaposleni[j].ime
        priimek2 = zaposleni[j].priimek
        print(f"\t{ime1} {priimek1} <--- {razlika} € ---> {ime2} {priimek2}")
````

    Napiši razred <b>Krizci_in_krozci</b>, ki simulira izvajanje igre "kržci in krožci".
    <br>Funkcije ki jih naj podpira:
    <br><b>__init__(self)</b>, ki inicializira celoten razred, ki se bo lahko uporabljal za igranje
    <br>
    <br><b>mark_X(self, row_index, column_index)</b>, ki označi X na željeno mesto
    <br>
    <br><b>mark_O(self, row_index, column_index)</b>, ki označi O na željeno mesto
    <br>
    <br><b>check_win(self)</b>, ki pregleda ali je kdo zmagal in napiše točno kdo je zmagal
    <br>
    <br><b>reset(self)</b>, ki pobriše že igrane X in O

````python
# Rešitev
class Krizci_in_krozci:
    def __init__(self):
        self.reset()
        
    def mark_X(self, row_index, column_index): # ki označi X na željeno mesto 
        self.board[row_index][column_index] = 'X'
        
    def mark_O(self, row_index, column_index): #ki označi O na željeno mesto
        self.board[row_index][column_index] = 'O'
        
    def check_win(self): # ki pregleda ali je kdo zmagal in napiše točno kdo je zmagal
        for line in range(3):
            # check rows
            combination = ''.join(self.board[line])
            self.check_combination(combination)
            
            # check columns
            combination = ''.join([self.board[el][line] for el in range(3)])
            self.check_combination(combination)
        
        # check cross 1
        combination = ''.join([self.board[el][el] for el in range(3)])
        self.check_combination(combination)
        
        # check cross 2
        combination = ''.join([self.board[-1-el][el] for el in range(3)])
        self.check_combination(combination)
    
    def check_combination(self, combination):
        if combination == 'XXX':
            print('Igralec, ki je označeval z X-em je zmagovalec!')
        elif combination == 'OOO':
            print('Igralec, ki je označeval z O-em je zmagovalec!')
    
    def reset(self):# ki pobriše že igrane X in O
        self.board = [[' ' for col in range(3)] for row in range(3)]
    
    def print_current_position(self): # ki izriše trenutno pozicijo igre v obliki kvadrata, kjer je vsaka celica označena s trenutnimi znaki (za izgled si poglej "Otput" primera spodaj). 
        for row in range(3):
            print(f'-------------')
            print('| ' + ' | '.join(self.board[row]) + ' |')
        print(f'-------------')
        
        
igra = Krizci_in_krozci()
plays = [(1,1),(0,0),
         (0,2),(1,0),
         (2,0)]

for i in range(len(plays)):
    if i%2==0:
        igra.mark_X(plays[i][0],plays[i][1])
    else:
        igra.mark_O(plays[i][0],plays[i][1])
    igra.print_current_position()
    igra.check_win()

igra.reset()
igra.print_current_position()
````
