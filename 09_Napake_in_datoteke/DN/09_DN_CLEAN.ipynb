````python
class Kalkulator:
    def __init__(self):
        self.operatorji = {'+': 'seštevanje',
                          '-': 'odštevanje',
                          '*': 'množenje',
                          '/': 'deljenje',
                          '**': 'potenciranje',
                          'q': 'zaključi kalkulator'}
        self.errors = 0
    
    def start(self):
        self.pomoc()
        while True:
            self.operator = input('Operator: ')
            if self.operator == 'q':
                print(f'število napak v času delovanja: {self.errors}')
                break
            self.x = input('x = ')
            self.y = input('y = ')
            new_errors = self.catch_errors()
            
            if new_errors:
                self.errors += new_errors
                print(f'[{self.x}] [{self.operator}] [{self.y}] = None')
                continue
            print(f'[{self.x}] [{self.operator}] [{self.y}] = {eval(self.x + self.operator + self.y)}')

    
    def pomoc(self):
        print('Operatorji:')
        for operator, description in self.operatorji.items():
            print(f'\t[{operator}] --> {description}')
    
    def catch_errors(self):
        current_errors = 0
        try: 
            float(self.x)
            float(self.y)
            if self.operator not in self.operatorji:
                raise ValueError('Operator je vnešen napačno')
        except Exception as e:
            print(f'Prišlo je do napake pri {type(e)}: {e}')
            current_errors += 1
        
        if current_errors == 0:
            return None
        else:
            return current_errors
        
````

Napišite funkcijo <code>izvedi_ukaze(ime_datoteke)</code>, ki prejme kot atribut ime datoteke, v kateri so v vsaki vrstici napisani ukazi, ki jih mora ta funkcija interpretirati. Ukazi se nanašajo na izrisovanje obrob likov. Primer datoteke je datoteka <code>test.txt</code>.

<br>Oblika ukaza: <code>[Lik], [Kam_izpisati], [Velikost]</code>

<br>Lik:
<br><code>trikotnik</code> --> izriše se naj obroba trikotnika
<br><code>kvadrat</code> --> izriše se naj obroba kvadrata



<br>Kam_izpisati:
<br><code>terminal</code> --> izriše se naj v terminalu
<br><code>ime_nove_datoteke</code> --> odpre se naj datoteko z imenom ime_nove_datoteke in vanj tudi izriše


<br>Velikost:
<br><code>stevilo_vrstic</code> --> koliko vrstic naj uporabi za izris (se upošteva tudi koliko je maksimalno znakov v vrstici --> brez '\n')


<br>Ob začetku preverjanja ukaza, naj v terminal izpiše kakšen je prebrani ukaz (lahko kar v listu, kot je narejeno v primerih), nato pa ga naj izvede. Če pride pri izvajanju do napake, naj program v terminalu napiše, da je prišlo do napake ter opis same napake.

<br>Za pomoč:
<br><code>split(znak)</code> --> razdeli string v list, kjer za razdelitev uporabi <code>znak</code> podan v oklepaju
<br><code>strip()</code> --> stringu obrže "prazne znake" kot so presledki, tabulatorji, ... Na čisto desni in levi strani stringa

<br>Razultat izvajanja so tudi ustvarjene datoteke

````python
def izvedi_ukaze(file):
    with open(file, 'r', encoding="utf-8") as f:
        lines = f.readlines()
    for line in lines:
        shape, location, size = line.strip('\n').split(',')
        shape, location, size = shape.strip(), location.strip(), size.strip()
        
        print(f'\nukaz: {[shape, location, size]}\n')
        geom = make_shape(shape, size)
        if not geom:
            continue
        
        if location == 'terminal':
            for line in geom:
                print(line)
        else:
            write_file(location, geom)
        
def make_shape(shape, size):
    try:
        size = int(size)
        assert size > 0, "Size should be above 0"
        assert shape in ['kvadrat', 'trikotnik'], "Shape not in predefined shapes"
    except Exception as e:
        print(f'\tError occured: {e}')
        # os.system(f'Error occured: {e}')
        return
    
    if shape == 'kvadrat':
        geom = make_square(size)
    else:
        geom = make_triangle(size)
    
    return geom
    
def make_square(size):
    if size == 1:
        return ['*']
        
    geom = ["*" + " " * (size - 2) + "*" for i in range(size)]
    geom[0] = '*' * size
    geom[-1] = '*' * size
    return geom
    
def make_triangle(size):
    if size == 1:
        return ['*']
    
    geom = ["*" + " " * (i - 1) + "*" for i in range(size)]
    geom[0] = '*'
    geom[-1] = '*' * size
    return geom

def write_file(file_name, lines):
    with open(file_name, 'w', encoding="utf-8") as f:
        for line in lines:
            f.write(line + '\n')
````
